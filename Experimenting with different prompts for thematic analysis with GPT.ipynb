```python
drive.mount('/content/drive',force_remount=True)

# After mounting, you can access files in Google Drive like this:

file_path = '/content/drive/My Drive/Colab Notebooks/cervical.txt'


def read_and_concatenate(file_path):
    try:
        with open(file_path, 'r') as file:
            # Read the entire file and replace newline characters with spaces
            concatenated_string = file.read().replace('\n', ' ')
            return concatenated_string
    except FileNotFoundError:
        return "File not found. Please provide a valid file path."


result = read_and_concatenate(file_path)

print( result)
```

This is the string I want you to experiment with editing. It provides the chatbot with a context for the question. Changing this should change the way it answers the questions.

```python
#context_prompt = "You are a qualitative researcher analysing free text samples for causal relationships. Identify all the causal relationships in the prompt, for each generate a pair: cause and effect"
context_prompt = "You are a qualitative researcher looking for common themes in patients' reviews of a clinical service. Identify the common themes in the prompt. Indicate how many patients are affected by each issue"
query = result
```

This cell sends a question from the file to the OpenAI API and prints the answer that is returned.

```python

from openai import OpenAI

client = OpenAI()

completion = client.chat.completions.create(
 #  model="gpt-3.5-turbo",
   temperature=0.7,  # Adjust the temperature as needed
   model= "gpt-4-0125-preview",
   messages=[
                {"role": "system", "content": context_prompt},
                {"role": "user", "content": query}
              ]
)

answer2 = completion.choices[0].message.content
formatted_text = textwrap.fill(answer2, width=70)

print(f'ChatGPT Answer: {formatted_text}\n')
```
